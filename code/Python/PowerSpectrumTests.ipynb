```python
import numpy as np
import scipy as sp
import scipy.integrate
import scipy.interpolate
%pylab inline
```

```python
# Set Up
# the space
l=5.; n=1000.
x = np.linspace(-l,l,n)
k = np.linspace(-l,l,n)
# the function
a=1.;b=1.;
f = lambda x: a*np.exp(-b*x*x)
```

```python
# first I compared numpy's FFT with an analytic FT and found that they were different
# using numpy's fft function
ft = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(f(x))))/np.sqrt(n)
# calculating an analytic FT
ftc = a*np.sqrt(np.pi/b)*np.exp(-k*k*np.pi*np.pi/b)
# plot stuff
ftp = plt.plot(k,ft,label='ft')
ftcp = plt.plot(k,ftc,label='ftc')
plt.legend()
```

```python
# even more concerningly, I found that numpy's fft changed based on the number of points, n
def plot_gauss_ft(n,c):
    l=5.;a=1.;b=1.;
    x = np.linspace(-l,l,n)
    k = np.linspace(-l,l,n)
    f = lambda x: a*np.exp(-b*x*x)
    ft = np.fft.fftshift(np.fft.fftn(np.fft.fftshift(f(x))))/np.sqrt(n)
    ftp = plt.plot(k,ft,label='n = {0}'.format(n))
    return ftp
for n,c in zip((50,100,500,1000),('b','g','r','o')):
    plot_gauss_ft(n,c)
plt.legend()    
```

```python
# Next I manually did a discrete FFT matches the analytic FT. 
# And it worked (after help from Josh). Yay!
l=5.; n=100.
a=1.;b=10.;
xx = np.linspace(-l,l,n)
kk = np.linspace(-n/(4*l),n/(4*l),n)
nx = np.arange(-n/2,n/2,dtype='float64')
nk = np.arange(-n/2,n/2,dtype='float64')
nxg,nkg = np.meshgrid(nx,nk)
ftd = np.sum(f(nxg*2*l/n).T*np.exp(-2*np.pi*1j*nxg*nkg/n)*2*l/n,axis=0)

ftc = a*np.sqrt(np.pi/b)*np.exp(-kk*kk*np.pi*np.pi/b)

# plot stuff
plt.plot(kk,ftc,label='ftc')
plt.plot(kk,ftd,label='ftd')
plt.xlim([-5,5])
plt.legend()
```
