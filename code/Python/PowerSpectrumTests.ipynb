```python
# ok. it's actually more complicated than I thought. I had been using lambda 
# functions, but I will need to be using arrays. hmmm.....
def discrete_ft(x,f):
    l = max(x); N = len(x) # assumes that x is from -l to l
    nx = N*x/(2.*l)
    nk = np.arange(-N/2.,N/2.,dtype='float64')
    nxg,nkg = np.meshgrid(nx,nk)
    fg,nkg = np.meshgrid(f,nk)
    ftd = np.sum(fg.T*np.exp(-2.*np.pi*1j*nxg*nkg/N)*2*l/N,axis=0)
    k = np.linspace(-N/(4*l),N/(4*l),N)
    return k,ftd
l=5.; n=50.
x = np.linspace(-l,l,n)
a=1.;b=1.;
f = a*np.exp(-b*x*x)
k,ftd = discrete_ft(x,f)
ftc = a*np.sqrt(np.pi/b)*np.exp(-k*k*np.pi*np.pi/b)

ftcp = plt.plot(k,ftc,label='ftc')
ftdp = plt.plot(k,ftd,label='ftd')
plt.xlim([-5,5]); plt.legend()
plt.show()
plt.plot(k,ftd-ftc,label='diff')
plt.xlim([-5,5]); plt.legend()
plt.show()
```

```python
# great. now on to the 2d case. I think this works now! :D
def discrete_ft_2d(x,y,f):
    lx = max(x); Nx = len(x) # assumes that x is from -l to l
    ly = max(y); Ny = len(y)
    nx = Nx*x/(2.*lx)
    kx = np.linspace(-Nx/(4*lx),Nx/(4*lx),Nx)
    ky = np.linspace(-Ny/(4*ly),Ny/(4*ly),Ny)
    xg,yg,kxg,kyg = np.meshgrid(x,y,kx,ky)
    fg = f(xg,yg) #a*np.exp(-b*(xg*xg+yg*yg))
    ftd = np.sum(fg*np.exp(-2*np.pi*1j*(kxg*xg+kyg*yg))*4*lx*ly/(Nx*Ny),axis=(0,1))
    return kx,ky,ftd
l=5.; n=100.
x = np.linspace(-l,l,n)
y = np.linspace(-l,l,n)
a=1.;b=1.;
f = lambda x,y: a*np.exp(-b*(x*x+y*y))
kx,ky,ftd = discrete_ft_2d(x,y,f)
kxg,kyg = np.meshgrid(kx,ky)
ftc = a*np.pi/b*np.exp(-(kxg*kxg+kyg*kyg)*np.pi*np.pi/b)
```
