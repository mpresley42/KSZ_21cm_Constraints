```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import gridspec
plt.rcParams['figure.figsize'] = (20.0, 16.0)
%pylab inline
```

## Functions from the Wavelets Module

```python
# Haar wavelet's mother wavelet function
mHaar = lambda x: np.piecewise(x, [x<0, x>=0, x>=0.5, x>=1], [0,1,-1,0])
# General Haar Function
Haar = lambda n,k,x: 2.**(n/2.)*mHaar((2.**n)*x-k)
```

```python
def wavelet_decomp(x,f,nvals,kvals):
    Cnk = np.zeros([len(nvals),len(kvals)])
    for ii in range(len(nvals)):
        for jj in range(len(kvals)):
            nkHaar = Haar(nvals[ii],kvals[jj],x)
            Cnk[ii,jj] = np.trapz(f*nkHaar,x=x)
    return Cnk
```

```python
def wavelet_recov(Cnk,nvals,kvals,x):
    f = np.zeros_like(x)
    for ii in range(len(nvals)):
        for jj in range(len(kvals)):
            nkHaar = Haar(nvals[ii],kvals[jj],x)
            f += Cnk[ii,jj]*nkHaar
    return f
```

```python
def build_intuition(f = lambda a,x: (2./np.pi)*np.arctan(a*x), avals=np.arange(0.5,3.0,0.5), xmax=10.,vmax=1):
    nmax = 10.; kmax = 10.
    dx = 0.1*2.**(-nmax)
    x = np.arange(-xmax,xmax,dx)
    nvals = np.arange(-nmax,nmax)
    kvals = np.arange(-kmax,kmax)

    fig, ax = plt.subplots(len(avals),2, sharex='col',figsize=(10,2*len(avals)))
    for ii,a in enumerate(avals):
        Cnk = wavelet_decomp(x,f(a,x),nvals,kvals)
        fr = wavelet_recov(Cnk,nvals,kvals,x)
        ax[ii,0].plot(x,f(a,x),label=r"$a = %s$"%a)
        ax[ii,0].plot(x,fr)
        ax[ii,0].set_ylabel(r"$f(x)$",fontsize=18)
        ax[ii,0].legend(loc='lower right',prop={'size':8})
        im = ax[ii,1].matshow(Cnk,origin='lower',vmin=-vmax,vmax=vmax,cmap='coolwarm')
        ax[ii,1].set_ylabel(r"$k$",fontsize=18)
        ax[ii,1].xaxis.set_ticks_position('bottom')
        ax[ii,1].set_aspect('equal',adjustable='box-forced')
        #ax[ii,0].set_aspect('equal',adjustable='box-forced')
    ax[-1,0].set_xlabel(r"$x$",fontsize=18)
    ax[-1,1].set_xlabel(r"$n$",fontsize=18)
    ax[0,0].set_title(r"$f(x)=\frac{2}{\pi}\arctan(ax)$",fontsize=18)
    ax[0,1].set_title(r"$C_{n,k}$",fontsize=18)

    cax = fig.add_axes([0.9, 0.1, 0.03, 0.8])
    fig.colorbar(im, cax=cax)
    #plt.show()
    return fig,ax
```

## Plotting Functions

```python
def plot_matrix(ax,M,x,y,xlabel='x',ylabel='y',grid=None,**kwargs):
    im = ax.matshow(M,origin='lower',extent=(x.min(),x.max()+1,y.min(),y.max()+1),**kwargs)
    ax.xaxis.set_ticks_position('bottom')
    if grid=='minor' or grid=='both': plt.minorticks_on()
    if grid=='major' or grid=='both': ax.grid(b=True, which='major', color='k', linestyle='-')
    if grid=='minor' or grid=='both': ax.grid(b=True, which='minor', color='k', linestyle='--',dashes=(2,5))
    ax.set_xlabel(r"$%s$"%xlabel,fontsize=18)
    ax.set_ylabel(r"$%s$"%ylabel,fontsize=18)
    plt.colorbar(im,ax=ax)
    return ax,im
```

```python
def wavelet_mapping(nlist,klist):
    fig = plt.figure(figsize=(8, 6)) 
    gs = gridspec.GridSpec(1,2,height_ratios=[3,1],width_ratios=[1.5,1])
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])
    ax0,im = plot_matrix(ax0,Cnk,nvals,kvals,xlabel='n',ylabel='k',grid='both',vmin=-1,vmax=1,cmap='coolwarm')
    ax0.scatter(nlist+0.5,klist+0.5,c='g')
    ax1.plot(x,ff,label='f(x)')
    for n,k in zip(nlist,klist):
        ax1.plot(x,Haar(n,k,x),label='n={0}; k={1}'.format(n,k))
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=8)
    fig.tight_layout()
    plt.show()
```

## Basic Decomposition

```python
nmax = 10.; kmax = 10.
dx = 0.1*2.**(-nmax)
x = np.arange(-10.,10.,dx)
ff = (2./np.pi)*np.arctan(0.5*x)
nvals = np.arange(-nmax,nmax)
kvals = np.arange(-kmax,kmax)
Cnk = wavelet_decomp(x,ff,nvals,kvals)

fig,ax = plt.subplots()
ax,im = plot_matrix(ax,Cnk,nvals,kvals,xlabel='n',ylabel='k',grid='both',vmin=-1,vmax=1,cmap='coolwarm')
plt.show()
```

## What does a step function look like?

```python
fstep = np.zeros_like(x)
fstep = np.where(x>0,-1,1)
stepCnk = wavelet_decomp(x,fstep,nvals,kvals)
fig,ax = plt.subplots()
ax,im = plot_matrix(ax,stepCnk,nvals,kvals,xlabel='n',ylabel='k',grid='both',vmin=-1.5,vmax=1.5,cmap='coolwarm')
plt.show()
```

## What happens if we increase the x range?

```python
nmax = 10.; kmax = 10.
dx = 0.1*2.**(-nmax)
nvals = np.arange(-nmax,nmax)
kvals = np.arange(-kmax,kmax)

xmaxvals = np.array([5.,6.,7.,8.,9.,10.])
fig, ax = plt.subplots(len(xmaxvals),2, sharex='col',figsize=(10,12))
for ii,xmax in enumerate(xmaxvals):
    x = np.arange(-xmax,xmax,dx)
    f = (2./np.pi)*np.arctan(0.5*x)
    Cnk = wavelet_decomp(x,f,nvals,kvals)
    fr = wavelet_recov(Cnk,nvals,kvals,x)
    ax[ii,0].plot(x,f,label=r"$xmax = %g$"%xmax)
    ax[ii,0].plot(x,fr)
    ax[ii,0].set_ylim([-1,1])
    ax[ii,0].set_ylabel(r"$f(x)$",fontsize=14)
    ax[ii,0].legend(loc='lower right',prop={'size':6})
    im = ax[ii,1].matshow(Cnk,origin='lower',vmin=-1,vmax=1,cmap='coolwarm',extent=(-nmax,nmax,-kmax,kmax))
    ax[ii,1].set_ylabel(r"$k$",fontsize=14)
    ax[ii,1].xaxis.set_ticks_position('bottom')
    ax[ii,1].set_aspect('equal',adjustable='box-forced')
    #ax[ii,0].set_aspect('equal',adjustable='box-forced')
ax[-1,0].set_xlabel(r"$x$",fontsize=14)
ax[-1,1].set_xlabel(r"$n$",fontsize=14)
ax[0,0].set_title(r"$f(x)=\frac{2}{\pi}\arctan(ax)$",fontsize=16)
ax[0,1].set_title(r"$C_{n,k}$",fontsize=16)

cax = fig.add_axes([0.9, 0.1, 0.03, 0.8])
fig.colorbar(im, cax=cax)
plt.show()
```
